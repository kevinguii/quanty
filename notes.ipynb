## SMA (Simple Moving Average)
- add a securities closing price over a given period that you want and average the values to get the moving averages
- reveal underlying trend
- long-term SMA have lag and smoother trend, short-term more sensitive

- when price crosses above SMA, it can signal a bullish entry
- when price falls below SMA, can signal bearish entry or reversal 
- if short-term signal goes above long-term signal (e.g. 50-day crosses 200-day), it could be a bullish signal
- trend-following strategy, mean reversion is opposite

## Mean Reversion
- idea that prices tend to revert back to their mean/historical average
	- if a price moves far above its average, its considered overbought, and will likely revert back down to its mean
	- if a price is far below its average, its oversold, likely to rise back
- in this case:
	- price > SMA = sell
	- price < SMA = buy

## Indicators of a Trend-Following or Mean-Reversion Strategy
- trending markets are usually trend-following
	- when the price moves in one direction consistently (e.g., tech stocks during a bull market)
	- momentum tends to continue, so better to follow the trend
- Range-bounded/sideways markets are better for mean reversion
	- price oscillates between a range without strong direction
	- price will often bounce back between the mean, better than chasing a trend

- strong SMA slope represents a trend, flat means a range
- ADX (Average Directional Index): measures trend strength, high is usually trend following, low is mean-reversion
- Bollinger Bands: prices hit outer bands of the market probably means good for mean-reversion

## MACD: Moving Average Convergence Divergence
- popular technical indicator to identify changes in strength, momentum of a trend
- includes MACD line, signal line, and histogram
	- MACD Line: difference between two EMAs of an asset price, typically 12-month EMA and 26-month EMA
	- signal line: nine-period EMA of MACD line, trigger for buying and selling
	- histogram: visual difference between MACD line and signal line

- bullish signal: when MACD line crosses above signal line
- bearish: when MACD crosses below signal line
- narrowing histogram can mean a crossover/reversal
- divergence: price of an asset moves in opposite direction of MACD indocator, signals potential change in trend momentum

CONS: prone to false signals, no risk management (stop-loss)

## Sharpe Ratio
- compares return of an investment with its risk
- numerator is risk-free rate of return, performance of investment over time
- denominator is std of returns over the same period (volatility and risk)
- higher is better, shows higher return per unit of risk
	- negative means its worse than treasury yield/risk-free rate
	- 1.0 is good, 2.0 is excellent, s&p around 0.73
