## Beneish M-Score

This is a statistical model that estimates the likelihood a company is manipulating earnings by using 8 financial ratios that look at things like revenue growth vs receivables, depreciation rates, margins, leverage, etc.

Variables:
1. DSRI: Days' sales in a receivable index
2. GMI: Gross margin index
3. AQI: Asset quality index
4. SGI: Sales growth index
5. DEPI: Depreciation index
6. SGAI: Sales and general and administrative expenses index
7. LVGI: Leverage index
8. TATA: Total accruals to total assets

Scoring Metric:


ex. sales grow 30% but acc receivable inc by 70%, mismatch could be a red flag, but more research should be done

## Sloan Accruals Anomaly

This is a market anomaly showing that firms with high accruals (earnings not backed by cashflow) tend to underperform. The logic is if earnings are driven mostly by accounting adjustments (accruals), it's most likely less sustainable.

A high accrual % of earnings likely means an overvalued stock. Lower % is higher quality.

ex. $500M in net income, but only $100M in operating cash flow. $400M accruals suggests earnings are accounting-driven, not cash-based

## Piotroski F-Score (Fundamental Health)

This is a 9-point scoring system used to measure financial strength, especially useful for valuing stocks.
Scores on profitability (ROA, CFO, earnings quality) is up to 4 points, leverage/liquidity (debt levels, liquidity, equity issuance) is up to 3 points, and operating efficiency (margin improvement, asset turnover) is up to 2 points.

A higher score means stronger and healthier fundamentals, less means weak and potentially distressed.

## Other Red Flags

- Unusual revenue recognition: booking sales early, channel stuffing (forcing distributors to take stuff)
- Receivables vs Sales Growth: receivables growing faster means customers aren't paying, fake sales risk
- Margins vs Peers: higher could mean manipulation
